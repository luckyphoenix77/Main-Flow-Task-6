```python
import pandas as pd

# Assuming you have already created the crosstab
b = pd.crosstab(df.Hypertension, df.Age)

# Find the maximum value in the Age columns
max_value_in_age = b.max().max()

# Find which row and column contains the maximum value
location = b[b == max_value_in_age].stack().idxmax()

print(f"The highest number is {max_value_in_age} in the row '{location[0]}' and column '{location[1]}'.")
```
